```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("I've been waiting to visit Alaska all my life.")
```

```python
classifier(
    ["I've been waiting to visit Alaska all my life.", "Its hell lot of fun!"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "The trip was all about the beauty of Alaska",
    candidate_labels=["glaciers", "husky", "football"],
)
```

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("On this trip, I experienced alot of")
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")
generator(
    "On this trip, I experienced alot of",
    max_length=30,
    num_return_sequences=2,
)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("I experienced alot of <mask> problems.", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Ravalika and I study at University of North Carolina in Charlotte.")
```

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I study?",
    context="My name is Ravalika and I study at University of North Carolina in Charlotte.",
)
```
